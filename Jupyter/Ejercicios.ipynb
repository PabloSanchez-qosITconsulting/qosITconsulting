## Ejercicio 1: Creación de una Clase Básica
Crea una clase llamada `Libro` que tenga los siguientes atributos:
- título
- autor
- año_publicacion
- disponible (booleano)

Implementa los métodos:
- `prestar()`: cambia disponible a False
- `devolver()`: cambia disponible a True
- `info()`: muestra toda la información del libro

```python
# Tu código aquí

class Libro:
    def __init__(self, titulo, autor, annoPublicacion,disponible):
        self.titulo = titulo
        self.autor = autor
        self.annoPublicacion = annoPublicacion
        self.disponible = True
        
        #Comprueba primero si el libro esta disponible
    def prestar(self):
        if self.disponible:
            self.disponible = False
            print("El libro se ha prestado correctamente")
        else:
            print("Libro ya se encuentra prestado")
        #Comprueba primero, si el libro se puede devolver
    def devolver(self):
        if not self.disponible:
            self.disponible = True
            print("El libro ha sido devuelto correctamente")
        else:
            print("El libro ya ha sido devuelto previamente.")
            
    def info(self):
        return print(f"Libro: {self.titulo} / Autor: {self.autor} / Año Publicacion: {self.annoPublicacion} / Disponibilidad: {self.disponible}")
    
    lib = Libro ('AA','BB',2020)
    lib.info()
    lib.prestar()
    lib.info()
    lib.devolver()
    lib.info
```

## Ejercicio 2: Herencia
Crea una clase `Vehiculo` con atributos básicos como `marca`, `modelo` y `año`.
Luego crea dos clases que hereden de `Vehiculo`:
- `Coche` (con atributo adicional `num_puertas`)
- `Moto` (con atributo adicional `cilindrada`)

Cada clase debe tener un método `mostrar_info()` que muestre sus atributos.

```python
# Tu código aquí

class Vehiculo:
    def __init__(self,marca,modelo,anno):
        self.marca=marca
        self.modelo=modelo
        self.anno = anno
    def mostrar_info(self):
        return "Marca: {self.marca} / Modelo: {self.modelo} / Año: {self.anno}"
    
class Coche(Vehiculo):
    def __init__(self, marca, modelo, anno,numPuertas):
        super().__init__(marca, modelo, anno)
        self.numPuertas = numPuertas
    def mostrar_info(self):
        return "Marca: {self.marca} / Modelo: {self.modelo} / Año: {self.anno} / Numero Puertas: {self.numPuertas}"

class Moto(Vehiculo):
    def __init__(self, marca, modelo, anno,cilindrada):
        super().__init__(marca, modelo, anno)
        self.cilindrada = cilindrada
    def mostrar_info(self):
        return "Marca: {self.marca} / Modelo: {self.modelo} / Año: {self.anno} / Cilindrada: {self.cilindrada}"
 
#Creacion vehiculos y mostrar información   
coch = Coche ('Peugot','A4',2015,7)
    
coch.mostrar_info()
    
mot = Moto ('Jog RR','R5',2019,49)
    
mot.mostrar_info()
```

## Ejercicio 3: Encapsulamiento
Crea una clase `CuentaBancaria` con:
- Atributos privados: `__saldo` y `__num_cuenta`
- Métodos:
  - `depositar(cantidad)`
  - `retirar(cantidad)`
  - `consultar_saldo()`

Asegúrate de que no se pueda retirar más dinero del disponible.

```python
# Tu código aquí

class CuentaBancaria:
    def __init__(self,saldo,numCuenta):
        self.__saldo = saldo
        self.__numCuenta = numCuenta
        
    def depositar(self,cantidad):
        self.__saldo = (self.__saldo + cantidad)
        print(f"Has depositado {cantidad}. Nuevo saldo: {self.__saldo}")
        
    def retirar(self,cantidad):
        if cantidad > self.__saldo:
            print("Tus fondos, estan a 0. Debes ingresar dinero")
        else:
            self.__saldo = (self.__saldo - cantidad)
            print(f"Has retirado {cantidad}. Nuevo saldo: {self.__saldo}")
        
    def consultar_saldo(self):
        print(f"Tu saldo actual es: {self.__saldo} unidades monetarias")
    
cuenta1 = CuentaBancaria(100,456)

#Ejemplos
cuenta1.consultar_saldo()
cuenta1.retirar(50)
cuenta1.consultar_saldo()
cuenta1.depositar(100)
cuenta1.consultar_saldo()
    
```

## Ejercicio 4: Polimorfismo
Crea una clase `FiguraGeometrica` con un método `calcular_area()`.
Luego crea las clases:
- `Rectangulo`
- `Circulo`
- `Triangulo`

Cada una debe implementar su propio método `calcular_area()`.

```python
# Tu código aquí
import math

class FiguraGeometrica:
    pass

class Rectangulo(FiguraGeometrica):
    def __init__(self, base,altura):
        self.base = base
        self.altura=altura
    def calcularArea(self):
        area = (self.base*self.altura)
        return area
        
        

class Circulo(FiguraGeometrica):
    def __init__(self, radio):
        self.radio=radio
    def calcularArea(self):
        area = math.pi*self.radio**2
        return area
        

class Triangulo(FiguraGeometrica):
    def __init__(self, base,altura):
        self.base = base
        self.altura = altura
    def calcularArea(self):
        area = (self.base*self.altura)/2
        return area
    
#Ejemplos
rectangulo = Rectangulo(5, 10)
print(f"Área del rectángulo: {rectangulo.calcularArea()}")

circulo = Circulo(7)
print(f"Área del círculo: {circulo.calcularArea()}")

triangulo = Triangulo(6, 8)
print(f"Área del triángulo: {triangulo.calcularArea()}")


```

## Ejercicio 5: Proyecto Final
Crea un sistema simple de gestión de una biblioteca usando POO.
Debe incluir:
- Clase `Biblioteca`
- Clase `Libro`
- Clase `Usuario`

Implementa métodos para:
- Agregar/eliminar libros
- Registrar usuarios
- Prestar/devolver libros
- Mostrar inventario
- Mostrar libros prestados

```python
# Tu código aquí
class Biblioteca:
    def __init__(self, nombre):
        self.nombre = nombre
        self.listaUsuarios = []
        self.listaLibros = []
    
    def registrar_usuario(self, usuario):
        if usuario not in self.listaUsuarios:
            self.listaUsuarios.append(usuario)
            return f"Usuario: {usuario.nombre} ha sido registrado correctamente"
        return f"El usuario {usuario.nombre} ya está registrado"
    
    def agregar_libro(self, libro):
        self.listaLibros.append(libro)
        return f"Libro '{libro.titulo}' agregado correctamente"
    
    def eliminar_libro(self, titulo):
        for libro in self.listaLibros:
            if libro.titulo == titulo and libro.disponible:
                self.listaLibros.remove(libro)
                return f"El libro '{titulo}' ha sido eliminado del stock correctamente"
        return "No se encontró el libro o está prestado en estos momentos"
    
    def mostrar_inventario(self):
        if not self.listaLibros:
            return "No hay libros en el inventario"
        
        inventario = "\nInventario Biblioteca:\n"
        for libro in self.listaLibros:
            inventario += f"{libro.info()}\n"
        return inventario
    
    def prestar_libro(self, titulo, usuario):
        for libro in self.listaLibros:
            if libro.titulo == titulo and libro.disponible:
                libro.disponible = False
                usuario.libros.append(libro)
                return f"El libro '{titulo}' se ha prestado correctamente a {usuario.nombre}"
        return "Libro no disponible o no encontrado"
    
    def mostrar_libros_prestados(self):
        prestados = [libro for libro in self.listaLibros if not libro.disponible]
        if not prestados:
            return "No hay libros prestados actualmente"
        
        resultado = "\nLibros prestados:\n"
        for libro in prestados:
            resultado += f"{libro.info()}\n"
        return resultado

class Usuario:
    def __init__(self, nombre, codigo_usuario):
        self.nombre = nombre
        self.codigo_usuario = codigo_usuario
        self.libros = []
    
    def devolver_libro(self, titulo, biblioteca):
        for libro in self.libros:
            if libro.titulo == titulo:
                libro.disponible = True
                self.libros.remove(libro)
                return f"El libro '{titulo}' ha sido devuelto correctamente"
        return "No tienes este libro prestado"
    
    def ver_libros_prestados(self):
        if not self.libros:
            return f"{self.nombre} no tiene libros prestados"
        
        resultado = f"\nLibros prestados a {self.nombre}:\n"
        for libro in self.libros:
            resultado += f"{libro.info()}\n"
        return resultado

class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponible = True
    
    def info(self):
        estado = "Disponible" if self.disponible else "Prestado"
        return f"Libro: {self.titulo} | Autor: {self.autor} | Estado: {estado}"
    
    # PRUEBAS
    
#Crear biblioteca
biblioteca = Biblioteca("Biblioteca Chipiona")

#Crear Libros
libro1 = Libro("Don Quijote", "Miguel de Cervantes")
libro2 = Libro("Cien años de soledad", "Gabriel García Márquez")

#Crear usuario
usuario1 = Usuario("Juan", "U001")

#Agregar libros
print(biblioteca.agregar_libro(libro1))
print(biblioteca.agregar_libro(libro2))

#Registrar usuario
print(biblioteca.registrar_usuario(usuario1))

#Mostrar inventario
print(biblioteca.mostrar_inventario())

#Prestar  libro
print(biblioteca.prestar_libro("Don Quijote", usuario1))

#Ver libros prestados
print(biblioteca.mostrar_libros_prestados())
print(usuario1.ver_libros_prestados())

#Devolver un libro
print(usuario1.devolver_libro("Don Quijote", biblioteca))


    

    

        
    

    
```
